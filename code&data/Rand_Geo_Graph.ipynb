```python
import numpy as np
import math

N = 100 # number of nodes
r = 0.2 # distance threshold
coordinates = np.random.rand(N,2)

print(coordinates)

graph = {}
for i in range(N):
    graph.update({i: []})
print(graph)
```

```python
for i in graph:
    for j in graph:
        if i != j:
            # coordinates for i node
            xi = coordinates[i][0] # value of column 0 in row i
            yi = coordinates[i][1] # value of column 1 in row i 
                                   # together they make coordinates of a node

            # coordinates for j node
            xj = coordinates[j][0]
            yj = coordinates[j][1]

            d = math.sqrt((xi-xj)**2 + (yi-yj)**2)
            if d <= r:
                graph[i].append(j)

graph
```

```python
# analysis
degrees = [len(neighbors) for neighbors in graph.values()]
degrees
```

```python
average_degrees = np.mean(degrees)
print(average_degrees)

# histogram
import matplotlib.pyplot as plt
plt.hist(degrees, bins=15, edgecolor='black')
plt.title("Histogram of degree distribution")
plt.xlabel("Degree / No of connections between nodes")
plt.ylabel("Number of nodes")
plt.show()
```

```python
# visualizing graph (nodes)

plt.figure(figsize=(8,8))
for i in graph:
    for j in graph[i]:
        if i < j:
            xval = [coordinates[i][0], coordinates[j][0]]
            yval = [coordinates[i][1], coordinates[j][1]]
            plt.plot(xval, yval, color='red', alpha =0.5, linewidth= 0.5)
x_nodes = coordinates[:, 0]
y_nodes = coordinates[:, 1]
plt.scatter(x_nodes, y_nodes, color='blue', s=15)
plt.title("Nodes graph")
plt.show()
```

```python
# wrapping this up in a single function

def rand_geo_graph(N,r):
    coordinates = np.random.rand(N,2)
    graph = {}
    for i in range(N):
        graph.update({i: []})
    
    for i in graph:
        for j in graph:
            if i != j:
                # coordinates for i node
                xi = coordinates[i][0] # value of column 0 in row i
                yi = coordinates[i][1] # value of column 1 in row i 
                                    # together they make coordinates of a node

                # coordinates for j node
                xj = coordinates[j][0]
                yj = coordinates[j][1]

                d = math.sqrt((xi-xj)**2 + (yi-yj)**2)
                if d <= r:
                    graph[i].append(j)
    
    degrees = [len(neighbors) for neighbors in graph.values()]
    average_degrees = np.mean(degrees)
    print(f'Average degrees : ', average_degrees)

    # histogram
    import matplotlib.pyplot as plt
    plt.hist(degrees, bins=15, edgecolor='black')
    plt.title("Histogram of degree distribution")
    plt.xlabel("Degree / No of connections between nodes")
    plt.ylabel("Number of nodes")
    plt.show()

    plt.figure(figsize=(8,8))
    for i in graph:
        for j in graph[i]:
            if i < j:
                xval = [coordinates[i][0], coordinates[j][0]]
                yval = [coordinates[i][1], coordinates[j][1]]
                plt.plot(xval, yval, color='red', alpha =0.5, linewidth= 0.5)
    x_nodes = coordinates[:, 0]
    y_nodes = coordinates[:, 1]
    plt.scatter(x_nodes, y_nodes, color='blue', s=15)
    plt.title("Nodes graph")
    plt.show()
    
```

```python
# N = 100, r = 0.2 - connected
rand_geo_graph(100, 0.2)
```

```python
# N = 100, r = 0.08 - not connected
rand_geo_graph(100, 0.08)
```

```python
# N = 1000, r = 0.08 - connected
rand_geo_graph(1000, 0.08)
```
