```python
import numpy as np
import random
from collections import deque

N = 100
k = 8
p = 0.2 # rewiring probability
```

```python
edges = []
for i in range(N):
    for d in range(1, k//2 + 1):
        r_neighbor = (i + d) % N
        edges.append((i, r_neighbor))

r_edges = []
for i, j in edges:
    if random.random() < p:
        new_j = random.randint(0, N - 1)
        while new_j == i or (i, new_j) in r_edges or (new_j, i) in r_edges:
            new_j = random.randint(0, N - 1)
        r_edges.append((i, new_j))
    else:
        r_edges.append((i, j))

graph = {i: [] for i in range(N)}
for i, j in r_edges:
    graph[i].append(j)
    graph[j].append(i)

print(graph)
```

```python
# graph construction
def node_distances(init_node, graph):
    distances = {node: -1 for node in graph} # -1 means not visited
    distances[init_node] = 0 # first node, distance from init_node to iself is 0
    
    queue = deque([init_node])
    while queue:
        curr_node = queue.popleft()         
        for neighbor in graph[curr_node]:
            if distances[neighbor] == -1:
                distances[neighbor] = distances[curr_node] + 1                
                queue.append(neighbor)
                
    return distances
```

```python
total_path_length = 0
for i in range(N):
    distances_from_i = node_distances(i, graph)
    for j in range(N):
        if i != j:
            total_path_length += distances_from_i[j]

average_shortest_path = total_path_length / (N * (N - 1))

print("Average Shortest Path:", average_shortest_path)
```

```python
def watts_strogatz(N,k,p):
    edges = []
    for i in range(N):
        for d in range(1, k//2 + 1):
            r_neighbor = (i + d) % N
            edges.append((i, r_neighbor))

    r_edges = []
    for i, j in edges:
        if random.random() < p:
            new_j = random.randint(0, N - 1)
            while new_j == i or (i, new_j) in r_edges or (new_j, i) in r_edges:
                new_j = random.randint(0, N - 1)
            r_edges.append((i, new_j))
        else:
            r_edges.append((i, j))

    graph = {i: [] for i in range(N)}
    for i, j in r_edges:
        graph[i].append(j)
        graph[j].append(i)

    def node_distances(init_node, graph):
        distances = {node: -1 for node in graph} # -1 means not visited
        distances[init_node] = 0 # first node, distance from init_node to iself is 0
        
        queue = deque([init_node])
        while queue:
            curr_node = queue.popleft()         
            for neighbor in graph[curr_node]:
                if distances[neighbor] == -1:
                    distances[neighbor] = distances[curr_node] + 1                
                    queue.append(neighbor)
                    
        return distances
    total_path_length = 0
    for i in range(N):
        distances_from_i = node_distances(i, graph)
        for j in range(N):
            if i != j:
                total_path_length += distances_from_i[j]

    average_shortest_path = total_path_length / (N * (N - 1))

    print("Average Shortest Path:", average_shortest_path)

watts_strogatz(100,8,0.2)
# works
```
